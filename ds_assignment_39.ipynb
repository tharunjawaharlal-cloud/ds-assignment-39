```python
"""Numpy Operations"""
```

```python
numeric_df = data.select_dtypes(include=np.number)
data_np = numeric_df.to_numpy()
```

```python
print("Min values:\n", np.nanmin(data_np, axis=0))
print("Max values:\n", np.nanmax(data_np, axis=0))
```

```python
print("\nPercentiles (25,50,75):\n",
      np.nanpercentile(data_np, [25, 50, 75], axis=0))
```

```python
mean_np = np.nanmean(data_np, axis=0)
std_np = np.nanstd(data_np, axis=0)
```

```python
print("\nNumPy Mean:\n", mean_np)
print("\nNumPy Std:\n", std_np)
```

```python
zscore_data = (data_np - mean_np) / std_np
print("\nZ-score sample (first 5 rows):\n", zscore_data[:5])
```
